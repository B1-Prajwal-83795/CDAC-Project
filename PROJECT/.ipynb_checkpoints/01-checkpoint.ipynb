```python
y = df['RainTomorrow']
x = df.drop('RainTomorrow',axis =1)
```

```python
from sklearn.preprocessing import StandardScaler
encoder=StandardScaler()
encoder.fit(x)
encoder=encoder.transform(x)
x_frame=pd.DataFrame(encoder,columns=x.columns)
```

```python
x_frame
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x_frame, y, random_state=124859 )
```

```python
from sklearn.linear_model import LogisticRegressionCV
model = LogisticRegressionCV(cv=10)
model.fit(x_train, y_train)
```

```python
def scores():
    from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
    y_pred=model.predict(x_test)
    y_true = y_test
    a = accuracy_score(y_pred, y_true)
    p=precision_score(y_pred, y_true)
    r=recall_score(y_pred, y_true)
    f1=f1_score(y_pred, y_true)
    # return a,p,r,f1
    print(f"accuracy_score={a},precision_score={p},recall_score={r},f1_score={f1}")
```

```python
scores()
```
